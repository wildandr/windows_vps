```python
import pandas as pd
```

```python
data = "datasets\GOLDm#\cleaned_data_M5.csv"

df = pd.read_csv(data)

df.tail()
```

```python
import matplotlib.pyplot as plt

# Convert the 'Time' column to datetime
df['Time'] = pd.to_datetime(df['Time'])

# Extract the date from the 'Time' column
df['Date'] = df['Time'].dt.date

# Group by the date and calculate the mean spread
daily_mean_spread = df.groupby('Date')['Spread'].mean()

# Plot the daily mean spread
plt.figure(figsize=(12, 6))
daily_mean_spread.plot()
plt.title('Daily Mean Spread')
plt.xlabel('Date')
plt.ylabel('Mean Spread')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import plotly.graph_objects as go

# Load the CSV file from the user's provided path
file_path = 'datasets/GOLDm#/cleaned_data_M5.csv'
data = pd.read_csv(file_path)

# Convert the Time column to datetime format
data['Time'] = pd.to_datetime(data['Time'])

# Creating the candlestick chart using Plotly
fig = go.Figure(data=[go.Candlestick(x=data['Time'],
                open=data['Open'],
                high=data['High'],
                low=data['Low'],
                close=data['Close'])])

# Update layout for better visibility
fig.update_layout(title='Candlestick Chart for GOLDm# Data',
                  xaxis_title='Time',
                  yaxis_title='Price',
                  xaxis_rangeslider_visible=False)

# Display the chart
fig.show()
```
